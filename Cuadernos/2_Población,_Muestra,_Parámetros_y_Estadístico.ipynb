La estadística es una disciplina esencial en el análisis de datos y la toma de decisiones en una amplia variedad de campos, desde la ciencia hasta los negocios. En el corazón de la estadística se encuentran conceptos clave como población, muestra, parámetros y estadísticos, que nos permiten comprender y describir el mundo que nos rodea de manera cuantitativa y significativa.

**Población: La Base de Todo**

<img src="https://github.com/jazaineam1/PYEF/blob/main/Imagenes/poblacion-mundial.png?raw=true"  width="400" height="300">


La población es el conjunto completo de elementos o individuos que estamos estudiando en un análisis estadístico. Puede ser tan amplia como "todos los seres humanos en el planeta" o tan específica como "todas las manzanas cosechadas en una plantación particular". En muchos casos, es impráctico o imposible analizar la población completa debido a su tamaño o inaccesibilidad. Aquí es donde entran en juego los conceptos de muestra, parámetros y estadísticos.

**Censo:**
Un censo es un proceso en el cual se recopila y registra información de todos los elementos de una población. En otras palabras, en un censo, se intenta medir o recopilar datos de todos los individuos o elementos que conforman una población específica. Los censos son útiles para obtener una imagen completa y precisa de una población, ya que no involucran la selección de muestras.

Sin embargo, realizar un censo puede ser costoso y consume mucho tiempo, especialmente en poblaciones grandes. Es por eso que, en muchos casos, se utiliza una muestra representativa en lugar de un censo completo para obtener información sobre la población. Las técnicas estadísticas se aplican para extraer conclusiones sobre la población basadas en la muestra.

**Muestra: Una Representación Accesible de la Población**

<img src="https://github.com/jazaineam1/PYEF/blob/main/Imagenes/muestra.png?raw=true"  width="400" height="300">

Dado que analizar una población completa es a menudo poco realista, recurrimos a muestras, que son subconjuntos más pequeños pero representativos de la población. Las muestras se seleccionan cuidadosamente para que reflejen las características clave de la población en estudio. Por ejemplo, si estamos interesados en conocer la opinión de los votantes sobre un tema político, podríamos encuestar a una muestra representativa de votantes registrados en lugar de encuestar a todos los votantes del país.

**Parámetros: Describiendo la Población**

Los parámetros son valores numéricos que describen las características de una población completa. Son cantidades fijas, pero a menudo desconocidas, ya que analizar una población completa es impráctico en la mayoría de los casos. Los parámetros son esenciales para comprender las propiedades de una población, pero generalmente se calculan utilizando información de una muestra representativa. Por ejemplo, si queremos conocer la edad promedio de todos los estudiantes en una universidad, el parámetro sería la edad promedio de **toda** la población de estudiantes.

**Estadísticos: Estimando Parámetros a partir de Muestras**

Los estadísticos son valores numéricos que se calculan a partir de una muestra y se utilizan para estimar los parámetros de la población. Los estadísticos proporcionan una imagen de las características de la muestra y se utilizan para hacer inferencias sobre la población en su conjunto. Por ejemplo, si tomamos una muestra de 100 estudiantes de una escuela y calculamos su edad promedio, este valor sería un estadístico que estima la edad promedio de **toda** la población de estudiantes en esa escuela.


**Ejemplo 1**

<img src="https://github.com/jazaineam1/PYEF/blob/main/Imagenes/altura.png?raw=true"  width="400" height="300">

Supongamos que estamos interesados en la altura promedio de todos los hombres adultos en un país. La población en este caso sería **todos** los hombres adultos del país. Debido a la dificultad de medir la altura de cada hombre adulto, tomamos una muestra de 500 hombres y calculamos su altura promedio. En este caso, la altura promedio de la muestra es un estadístico, mientras que la altura promedio de **todos** los hombres adultos del país sería el parámetro.

```python
import numpy as np
import pandas as pd
# Generar una población de alturas de hombres adultos (en cm)
poblacion = np.round(np.random.normal(175, 7, 1000000) ) # Media = 175, Desviación estándar = 7

# Tomar una muestra aleatoria de la población
tamano_muestra = 500
muestra = np.random.choice(poblacion, size=tamano_muestra)

# Calcular la media de la muestra (estadístico)
media_muestra = np.mean(muestra)
print("Media de la muestra:", media_muestra)

# Calcular la media de la población (parámetro)
media_poblacion = np.mean(poblacion)
print("Media de la población:", media_poblacion)
```

```python
import numpy as np
import pandas as pd

# Generar coordenadas negativas aleatorias
np.random.seed(42)  # Fijar la semilla para reproducibilidad
aleatorio = np.random.uniform(0, 1, 100)

# Generar edades aleatorias (variable cuantitativa)
edades = np.random.randint(18, 65, 100)

# Generar género aleatoriamente (variable cualitativa)
sexo = np.random.choice(['Masculino', 'Femenino'], size=100)

# Crear un DataFrame con los datos
data = pd.DataFrame({ 'Edad': edades, 'Sexo': sexo})
display(data)
# Calcular media, varianza, desviación estándar y proporción
media_edad = data['Edad'].mean()
varianza_edad = data['Edad'].var()
desviacion_estandar_edad = data['Edad'].std()
proporcion_genero = data['Sexo'].value_counts(normalize=True)

print("Media de Edad:", media_edad)
print("Varianza de Edad:", varianza_edad)
print("Desviación Estándar de Edad:", desviacion_estandar_edad)
print("Proporción de Género:\n", proporcion_genero)
```

```python
data['Aleatorio']=aleatorio
data
```

```python
muestra=data.sort_values('Aleatorio').head(40)
muestra
```
