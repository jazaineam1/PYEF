# Introducción a Seaborn

Seaborn es una libreria de python para la visualiazación de datos basada en matplotlib. Esta librería esta hecha para construir gráficos estadísticos y tiene la ventaja de estar estrechamente integrada a la librería de pandas.

Como siempre, antes de comenzar a producir cualquier gráfico, es necesario tener un amplio entendimiento de la estructura de los datos, así como también, tener el conjunto de datos preprocesado y listo para trabajar. Dentro de las funciones que ofrece Seaborn se destacan las siguientes:

- Esta orientada al trabajo con dataframes de pandas, lo que permite examminar relaciones entre múltiples variables
- Tiene un gran soporte para el manejo de variables categóricas y para pintar observaciones y estadísticas agregadas
- Opciones para visualizar distribuciones univaridas y bivariadas
- Estimación automática como por ejemplo el trazado de modelos de regresión lineal
- Vistas sobre la estructura general de conjuntos de datos
- Abstracciones de alto nivel para estructurar cuadrículas de múltiples gráficos
- Control del estilo con varios temas incorporados
- Herramientas para elegir paletas de colores

## Tablas de frecuencias y gráficas para variables cualitativas
La estadística descriptiva se encarga de organizar datos con el fin de mostrar la distribución general de éstos y el lugar en donde tienden a concentrarse, además de señalar valores de datos poco usuales o extremos. El primer procedimiento que se emplea para organizar y resumir un conjunto de datos es una tabla de frecuencias.

<img src="https://antoninoapp.shinyapps.io/semana2/_w_d6dafdb0/www/ana.png"  width="150" height="130">

Si hablamos de frecuencia estadística, podemos definirlo como el número de veces que se repite un dato en una muestra estadística.

Para definir correctamente las frecuencias necesitamos diferenciar entre dos tipos o dos maneras de representarlas:

**Frecuencia Absoluta:** es la definición que entenderíamos habitualmente como frecuencia estadística. Básicamente es el conteo de operaciones que se repiten en una muestra o experimento.



````python
f_ab=dfe['programa'].value_counts()
f_ab
````

**Frecuencia Relativa:** se trata simplemente del cociente entre la frecuencia absoluta (calculada según las indicaciones del punto anterior) y el tamaño de la muestra o el número de veces que se haya realizado un experimento.

````python
f_re=f_ab/f_ab.sum()
f_re*100
````

````python
print('Frecuencia absoluta acumulada:','\n',f_ab.cumsum())
print('\n\n','Frecuencia relativa acumulada:','\n',f_re.cumsum())
````

### Diagrama de barras
En ella, las clases o categorías se representan en el eje horizontal y la frecuencia de clase en el eje vertical. Las frecuencias de clase son proporcionales a las alturas de las barras.

````python
sns.barplot(x=f_ab.index, y=f_ab.values)
plt.xticks(rotation=90)
plt.show()
````

## Diagrama de torta
Gráfica que muestra la parte o porcentaje que representa cada clase del total, de números de frecuencia.

````python
plt.pie(f_ab,labels=f_ab.index, autopct='%.0f%%')
plt.show()
````

### Gráfica de líneas
Cuando una variable cuantitativa se registra en el tiempo a intervalos igualmente espaciados (por ejemplo: diario, semanal, mensual, trimestral o anual) el conjunto de datos forma una serie de tiempo.

Para esto usaremos el precio de la acción de apple desde el 17 de febrero del 2015 hasta el 16 de febrero de 2017.

````python
df_apple =pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
df_apple.columns
ax=sns.relplot(x='Date', y='AAPL.Close', data=df_apple, kind='line')
ax.set_xticklabels(rotation=30)
plt.show()
````

````python
df_apple['date2'] = pd.to_datetime(df_apple['Date'], format = '%Y-%m-%d')
ax=sns.relplot(x='date2', y='AAPL.Close', data=df_apple, kind='line')
ax.set_xticklabels(rotation=30)
plt.show()
````

## Diagrama de caja

Se trata de una representación gráfica sencilla que no necesita un número elevado de valores para su construcción. Sirve para visualizar tanto la dispersión como la forma de una variable. Asimismo, es especialmente útil para comparar diferentes distribuciones de manera simultánea.

Como dice su nombre, consta de una caja, donde la misma representa el cincuenta por ciento central de la distribución (va de
Q1
 o primer cuartil a Q3
 o tercer cuartil), y la línea situada en el interior de la caja es la mediana.

En este gráfico, Q1
 recibe el nombre de bisagra inferior y Q3
 bisagra superior. Los extremos inferiores y superiores de los segmentos (también llamados bigotes) delimitan lo que se denomina valores normales y coinciden, respectivamente, con el mínimo y el máximo de los valores una vez excluidos los candidatos a valores anómalos. Los candidatos a valores anómalos se etiquetan como atípicos y coinciden con aquellas observaciones que se encuentran fuera del intervalo (LI,LS) , donde:

 $LI=Q_1−1.5\times R_I,LS=Q_3+1.5\times R_I$

 con $R_i$ llamado el rango intercuartilico definido como $R_I$=$Q_3−Q_1$

Seaborn dispone de la función __relplot__, la cual presenta variadas e interesantes parametrizaciones.

### Otras opciones de visualización de gráficas dispersas

Varias dimensiones se pueden visualizar en una gráfica de dispersión
- $x$ variable numérica (puede tener múltiples valores)
- $y$ variable numérica
- $hue$ variable categórica (agrupa)
- $size$ diferentes tamaños de la burbuja, variable numérica
- $style$ pone marcas diferentes, variable categórica (agrupa)
- $col$ organiza filtros por columna, variable categórica (filtra)
- $row$ organiza filtros por fila, variable categórica (filtra)

#### Ejemplo





```python
# Mostrar varias dimensiones con el dataframe de Breast cancer.
```
